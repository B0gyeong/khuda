```python
import pandas as pd
import numpy as np

feature_with_weather = pd.read_csv('/content/drive/MyDrive/KHUDA/카페데이터/최종/feature_by_weather_without_iceamericano.csv', encoding = 'cp949')
feature_with_avg_weather = pd.read_csv('/content/drive/MyDrive/KHUDA/카페데이터/최종/ratio_features_without_iceamericano.csv', encoding = 'cp949')
feature = pd.read_csv('/content/drive/MyDrive/KHUDA/카페데이터/최종/features_without_iceamericano.csv', encoding = 'cp949')
```

```python
feature_weather = feature_with_weather.drop(['기온', '흐림', '비소나기눈'], axis=1)
print(feature_weather.shape)
feature_weather.head()
```

```python
drink_feature= feature.drop(['Unnamed: 0'], axis=1)
print(drink_feature.shape)
drink_feature.head()
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

def find_similarity(numpy_feature, drink_feature):
  answer = []
  for i in range(37):
    drink_feature_solo = drink_feature.iloc[i,1:]
    numpy_drink_feature = drink_feature_solo.to_numpy()
    numpy_drink_feature = numpy_drink_feature.astype(float)
    value = euclidean_distance(numpy_feature, numpy_drink_feature)
    answer.append(value)
  return answer

def get_min_indices(lst, n):
    min_indices = []
    sorted_indices = sorted(range(len(lst)), key=lambda i: lst[i])
    for i in range(n):
        min_indices.append(sorted_indices[i])
    return min_indices

def get_items(min_indices):
  min_items = []
  for i in range(len(min_indices)):
    min_items.append(drink_feature.loc[min_indices[i], 'items'])
  return min_items

def recommend_drink_by_weather(feature_weather,drink_feature):
  lst=[]
  for i in range(len(feature_weather)):
    numpy_feature = feature_weather.iloc[i]
    value = find_similarity(numpy_feature, drink_feature)
    lst.append(value)
  return lst
```

```python
recommend = recommend_drink_by_weather(feature_weather,drink_feature)
```

```python
indices = []
for i in range(len(recommend)):
  value = get_min_indices(recommend[i], 3)
  indices.append(value)

indices
```

```python
min_items = []
for j in range(len(indices)):
  value = get_items(indices[j])
  min_items.append(value)

for i in range(16):
  print(i, min_items[i])
```

표준화한 데이터에 대한 음료 추천

```python
feature_avg_weather = feature_with_avg_weather.drop(['기온', '흐림', '비소나기눈'], axis=1)
print(feature_avg_weather.shape)
feature_avg_weather.head()
```

```python
avg_recommend = recommend_drink_by_weather(feature_avg_weather,drink_feature)
```

```python
avg_indices = []
for i in range(len(avg_recommend)):
  value = get_min_indices(avg_recommend[i], 6)
  avg_indices.append(value)

avg_indices
```

```python
avg_min_items = []
for j in range(len(avg_indices)):
  value = get_items(avg_indices[j])
  avg_min_items.append(value)

for i in range(16):
  print(i, avg_min_items[i])
```

원래 자료에서 상위 3개와 표준화 자료 상위 3개 (총6개) 추천

```python
# min_items, avg_min_items

for i in range(len(avg_min_items)):
  count = 0
  for j in range(len(avg_min_items[i])):
    if avg_min_items[i][j] not in min_items[i]:
      min_items[i].append(avg_min_items[i][j])
      count = count +1
    else:
      continue;
    if count == 3:
      break;
```

```python
for i in range(16):
  print(i, min_items[i])
```
